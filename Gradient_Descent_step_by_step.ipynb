```python
from jupyter_server.serverapp import list_running_servers
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y= make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
# lets apply gradient decent assuming slope is constant m = 78.35
# and let  assume the starting value for intercept b=0
y_pred = ((78.35 * x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='blue',label='b = o')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
# lets take learnig rate = 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
#calculating the new intercept
b =  b - step_size
b
```

```python
y_pred1 = ((78.35 * x)+b).reshape(4)


plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='green',label='OLS')
plt.plot(x,y_pred1,color='blue',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='red',label='b = 0')
plt.legend()
plt.show()
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='green',label='OLS')
plt.plot(x,y_pred2,color='blue',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='pink',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='red',label='b = 0')
plt.legend()
plt.show()
```
